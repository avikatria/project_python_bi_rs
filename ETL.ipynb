<h3>Menghapus Duplikat</h3>

```python
def drop_duplicates_and_report(df, key_columns=None, name=""):
    before = len(df)
    if key_columns:
        df = df.drop_duplicates(subset=key_columns)
        print(f"{name} → Hapus duplikat berdasarkan kolom {key_columns}: {before - len(df)} baris")
    else:
        df = df.drop_duplicates()
        print(f"{name} → Hapus duplikat seluruh baris: {before - len(df)} baris")
    return df

# Contoh penggunaan:
patient = drop_duplicates_and_report(patient, ['patient_id'], "patient")
doctor = drop_duplicates_and_report(doctor, ['doctor_id'], "doctor")
department =drop_duplicates_and_report(department, ['department_id'], "department")
nurse = drop_duplicates_and_report(nurse, ['nurse_id'], "nurse")
appointment = drop_duplicates_and_report(appointment, ['appointment_id'], "appointment")
prescription = drop_duplicates_and_report(prescription, ['prescription_id'], "prescription")
billing = drop_duplicates_and_report(billing, ['bill_id'], "billing")
room_assignment = drop_duplicates_and_report(room_assignment, ['assignment_id'], "room_assignment")
room = drop_duplicates_and_report(room, ['room_id'], "room")
department = drop_duplicates_and_report(department, ['department_id'], "department")
medicine = drop_duplicates_and_report(medicine, ['medicine_id'], "medicine")
medical_record = drop_duplicates_and_report(medical_record, ['record_id'], "medical_record")
staff = drop_duplicates_and_report(staff, ['staff_id'], "staff")
pharmacy = drop_duplicates_and_report(pharmacy, ['pharmacy_id'], "pharmacy")
cleaning_service = drop_duplicates_and_report(cleaning_service, ['service_id'], "cleaning_service")
ambulance_log = drop_duplicates_and_report(ambulance_log, ['log_id'], "ambulance_log")

```

```python
import re

# 1. Nama → kapitalisasi judul, buang spasi
patient['first_name'] = patient['first_name'].str.strip().str.title()
patient['last_name'] = patient['last_name'].str.strip().str.title()
doctor['first_name'] = doctor['first_name'].str.strip().str.title()
doctor['last_name'] = doctor['last_name'].str.strip().str.title()
staff['first_name'] = staff['first_name'].str.strip().str.title()
staff['last_name'] = staff['last_name'].str.strip().str.title()

# 2. Gender → uppercase, normalisasi ke M/F
def normalize_gender(g):
    g = str(g).strip().upper()
    return 'M' if g in ['M', 'MALE', 'L'] else ('F' if g in ['F', 'FEMALE', 'P'] else None)

patient['gender'] = patient['gender'].apply(normalize_gender)


# 3. Email → lowercase & strip
staff['email'] = staff['email'].str.strip().str.lower()
doctor['email'] = doctor['email'].str.strip().str.lower()
staff['email'] = staff['email'].str.strip().str.lower()

# 4. Validasi format email
def is_valid_email(email):
    if pd.isnull(email): return False
    return bool(re.match(r"[^@]+@[^@]+\.[^@]+", email))

# Contoh cek yang tidak valid:
invalid_staff_email = staff[~staff['email'].apply(is_valid_email)]
print(f"staff → Email tidak valid: {len(invalid_staff_email)}")

# 5. Telepon → hapus spasi & karakter non-digit (opsional)
for df in [doctor, staff, patient]:
    if 'contact_number' in df.columns:
        df['contact_number'] = df['contact_number'].astype(str).str.replace(r'\D', '', regex=True)

# 6. Nama departemen → title case
department['depart_name'] = department['department_name'].str.strip().str.title()
room['room_number'] = room['room_number'].astype(str).str.strip().str.upper()

print("format teks dan nilai kategorikal distandarkan.")
```

<h3>Cek dan Koreksi Relasi Antar Tabel</h3>

```python
def cek_relasi(df_asal, kolom_fk, df_target, kolom_pk):
    missing = ~df_asal[kolom_fk].isin(df_target[kolom_pk])
    if missing.sum() > 0:
        print(f"❌ {missing.sum()} data di kolom '{kolom_fk}' tidak ditemukan di '{kolom_pk}'")
        return df_asal[missing]  # tampilkan baris yang bermasalah
    else:
        print(f"✅ Semua nilai di kolom '{kolom_fk}' cocok dengan '{kolom_pk}'")
        return None
```

```python
from datetime import datetime

# Pastikan kolom date_of_birth adalah datetime
patient['date_of_birth'] = pd.to_datetime(patient['date_of_birth'])

# Tanggal acuan (misalnya hari ini)
today = pd.to_datetime('today')

# Fungsi hitung umur
def calculate_age(birth_date, reference_date):
    return reference_date.year - birth_date.year - (
        (reference_date.month, reference_date.day) < (birth_date.month, birth_date.day)
    )

# Terapkan ke semua baris
patient['age'] = patient['date_of_birth'].apply(lambda dob: calculate_age(dob, today))
patient.head()
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Boolean, Date, ForeignKey
from sqlalchemy.orm import declarative_base
from sqlalchemy.orm import relationship, sessionmaker

# Buat koneksi
engine_dwh = create_engine('postgresql://postgres:1234567@localhost:5432/dw_hospital_ms')
Base = declarative_base()

# Tabel Dim Date
class DimDate(Base):
    __tablename__ = 'dim_date'
    date_id = Column(Integer, primary_key=True)
    payment_date = Column(Date)
    payment_month = Column(Integer)
    payment_year = Column(Integer)

# Tabel Dim Bill
class DimBill(Base):
    __tablename__ = 'dim_bill'
    bill_id = Column(String, primary_key=True)
    insurance_provider = Column(String)
    paid_late = Column(Boolean)

# Tabel Fact Billing
class FactBilling(Base):
    __tablename__ = 'fact_billing'
    fact_id = Column(Integer, primary_key=True)
    date_id = Column(Integer, ForeignKey('dim_date.date_id'))
    bill_id = Column(String, ForeignKey('dim_bill.bill_id'))

    # Opsional: relasi (biar bisa join otomatis kalau pakai ORM)
    date = relationship('DimDate')
    bill = relationship('DimBill')
```

```python
from sqlalchemy import Column, String, Float, Integer, ForeignKey, create_engine

Base = declarative_base()

class DimMedicine(Base):
    __tablename__ = 'dim_medicine'
    medicine_id = Column(String, primary_key=True)
    name = Column(String)
    dosage_value = Column(Float)
    dosage_unit = Column(String)
    type = Column(String)

class FactStock(Base):
    __tablename__ = 'fact_stock'
    fact_id = Column(Integer, primary_key=True)
    medicine_id = Column(String, ForeignKey('dim_medicine.medicine_id'))
    remaining_quantity = Column(Float)

    # Opsional relasi
    medicine = relationship('DimMedicine')

# Buat engine dan tabel
engine_dwh = create_engine('postgresql://postgres:1234567@localhost:5432/dw_hospital_ms')
Base.metadata.create_all(engine_dwh)
```

<h3>Pertanyaan 3 : Jenis Kamar Apa yang Paling Sering Digunakan?</h3>

<h4>Buat tabelnya di postgre</h4>

```python
from sqlalchemy import create_engine, Column, Integer, String, Date, ForeignKey
from sqlalchemy.orm import declarative_base, relationship

Base = declarative_base()

class DimRoom(Base):
    __tablename__ = 'dim_room'
    room_id = Column(Integer, primary_key=True)
    room_type_id = Column(Integer)
    room_class = Column(String)

class DimDateRoom(Base):
    __tablename__ = 'dim_date_room'
    date_id = Column(String, primary_key=True)
    assignment_date = Column(Date)
    assignment_month = Column(Integer)
    assignment_year = Column(Integer)

class FactRoomUsage(Base):
    __tablename__ = 'fact_room_usage'
    fact_id = Column(Integer, primary_key=True)
    room_id = Column(Integer, ForeignKey('dim_room.room_id'))
    date_id = Column(String, ForeignKey('dim_date_room.date_id'))
    usage_count = Column(Integer)
    length_of_stay_total = Column(Integer)

    # Optional: definisikan relasi ORM
    room = relationship('DimRoom')
    date = relationship('DimDateRoom')
    
engine_dwh = create_engine('postgresql://postgres:1234567@localhost:5432/dw_hospital_ms')
Base.metadata.create_all(engine_dwh)
```

<h4>Masukin datanya</h4>

```python
dim_room.to_sql('dim_room',con=engine_dwh, if_exists='append', index=False)
dim_date_room.to_sql('dim_date_room', con=engine_dwh, if_exists='append', index=False)
fact_room_usage.to_sql('fact_room_usage', con=engine_dwh, if_exists='append', index=False)
```

<h3>Pertanyaan 4 : Bagaimana Tren Penggunaan Layanan Ambulan?</h3>

<h4>Buat tabelnya di postgre</h4>

```python
from sqlalchemy import Column, Integer, String, Numeric, Boolean, ForeignKey, DateTime
from sqlalchemy.orm import declarative_base

Base = declarative_base()

class DimDateAmbulan(Base):
    __tablename__ = 'dim_date_ambulan'
    date_id = Column(Integer, primary_key=True)
    full_date = Column(DateTime)
    month = Column(Integer)
    year = Column(Integer)

class DimAmbulance(Base):
    __tablename__ = 'dim_ambulance'
    ambulance_id = Column(Integer, primary_key=True)
    ambulance_number = Column(String)
    availability = Column(String)

class FactAmbulanceUsage(Base):
    __tablename__ = 'fact_ambulance_usage'
    fact_id = Column(Integer, primary_key=True)
    ambulance_id = Column(Integer, ForeignKey('dim_ambulance.ambulance_id'))
    date_id = Column(Integer, ForeignKey('dim_date_ambulan.date_id'))
    journey_duration_minutes = Column(Numeric)

    # Opsional relasi
    date = relationship('DimDateAmbulan')
    ambulance = relationship ('DimAmbulance')

engine_dwh = create_engine('postgresql://postgres:1234567@localhost:5432/dw_hospital_ms')
Base.metadata.create_all(engine_dwh)
```

```python
dim_date_ambulan.to_sql('dim_date_ambulan', con=engine_dwh, if_exists='append', index=False)
dim_ambulance.to_sql('dim_ambulance', con=engine_dwh, if_exists='append', index=False)
fact_ambulance_usage.to_sql('fact_ambulance_usage', con=engine_dwh, if_exists='append', index=False)
```

<h3>Pertanyaan 5 : Berapa Estimasi Penyakit yang Akan Datang Pada Bulan Depan?</h3>

<h4>Buat tabelnya di postgre</h4>

```python
from sqlalchemy import Column, Integer, String, Text, Date, ForeignKey, create_engine
from sqlalchemy.orm import declarative_base, relationship

Base = declarative_base()

# DIMENSIONS
class DimDateDokpas(Base):
    __tablename__ = 'dim_date_dokpas'
    date_id = Column(Integer, primary_key=True)  # YYYYMMDD
    full_date = Column(Date)
    month = Column(Integer)
    year = Column(Integer)

class DimDoctor(Base):
    __tablename__ = 'dim_doctor'
    doctor_dim_id = Column(Integer, primary_key=True)
    doctor_id = Column(Integer, unique=True)
    specialty = Column(String) 

# FACTS
class FactDiagnosis(Base):
    __tablename__ = 'fact_diagnosis'
    fact_id = Column(Integer, primary_key=True)
    date_id = Column(Integer, ForeignKey('dim_date_dokpas.date_id'))
    diagnosis = Column(Text)
    diagnosis_count = Column(Integer)

    date = relationship('DimDateDokpas')

class FactDoctorPatient(Base):
    __tablename__ = 'fact_doctor_patient'
    fact_id = Column(Integer, primary_key=True)
    doctor_dim_id = Column(Integer, ForeignKey('dim_doctor.doctor_dim_id'))
    date_id = Column(Integer, ForeignKey('dim_date_dokpas.date_id'))
    total_patients = Column(Integer)

    date = relationship('DimDateDokpas')
    doctor = relationship('DimDoctor')

# ENGINE & CREATE
engine_dwh = create_engine('postgresql://postgres:1234567@localhost:5432/dw_hospital_ms')
Base.metadata.create_all(engine_dwh)
```
